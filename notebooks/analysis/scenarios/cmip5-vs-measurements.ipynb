```python
import xarray as xr
import pandas as pd

%matplotlib inline
```

```python
# times are not readable by xarray ('years' are not recognized)
ds_low = xr.open_dataset('../../../data/knmi/cmip5/SeaLevelPerc_KNMI14_rcp45.nc', decode_times=False)
ds_high = xr.open_dataset('../../../data/knmi/cmip5/SeaLevelPerc_KNMI14_rcp85.nc', decode_times=False)
```

```python
df_low = ds_low.to_dataframe().reset_index()
df_low['scenario'] = 'rcp45'
df_high = ds_high.to_dataframe().reset_index()
df_high['scenario'] = 'rcp85'

df = pd.concat([df_low, df_high])
```

```python
df.to_csv('../../../data/knmi/cmip5/SeaLevelPerc_KNMI14.csv')
```
