```python
import pathlib

import xarray as xr

import geopandas as gpd
import pandas as pd
```

```python
data_dir = pathlib.Path('~/data/gtsm/hourly').expanduser()
```

```python
paths = list(data_dir.glob('*.nc'))
```

```python
ds = xr.open_dataset(paths[0])
ds
```

```python
geometry = gpd.points_from_xy(ds.station_x_coordinate, ds.station_y_coordinate)
```

```python
gpd.GeoDataFrame(data=dict(station=ds.stations), geometry=geometry)
```
